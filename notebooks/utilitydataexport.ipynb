```python
import geopandas as gpd
import pandas as pd
import requests
from time import sleep
```

```python
metros = gpd.read_file('../data/metro/tl_2020_us_uac10.shp')
n = 206
list_metros = [metros[i:i+n] for i in range(0,metros.shape[0],n)]
```

```python
#Endpoint for API with key attached
url = 'https://api.openei.org/utility_rates?version=3&format=json&api_key=y6C2p5K9VoXH2qYEyxlkJaNco4uEKVnyhR5BirL4'
```

```python
x = []
name = iter(range(8,50))
for metrodf in list_metros[8:]:
    utility_region = pd.DataFrame()
    for index, row in metrodf.iterrows():
        params = {'sector':'Residential',          
             'lat':row.INTPTLAT10,
             'lon':row.INTPTLON10
             }
        response = requests.get(url, params = params)
        print(str(index)+' - ' +str(response))
        if len(response.json()['items']):
            df = pd.json_normalize(response.json()['items'])
            df['GEOID10'] = row['GEOID10']
            try:
                df = df[df.enddate.isna()]
            except:
                x.append(index)
            utility_region = utility_region.append(df.loc[:,['GEOID10','utility','eiaid']].drop_duplicates())
        else:
            utility_region = utility_region.append({'region_name': row['GEOID10'],'utility':'No','eiaid':'No'}, ignore_index=True)
    utility_region.to_csv('../data/rates' + str(next(name)) + '.csv')
    sleep(180)
```
