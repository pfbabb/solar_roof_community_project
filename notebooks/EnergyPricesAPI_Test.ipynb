```python
x = []
utility_region = pd.DataFrame()
for index, row in metros.iloc[0:10].iterrows():
    params = {'sector':'Residential',          
         'lat':row.INTPTLAT10,
         'lon':row.INTPTLON10
         }
    response = requests.get(url, params = params)
    print(str(index)+' - ' +str(response))
    if len(response.json()['items']):
        df = pd.json_normalize(response.json()['items'])
        df['GEOID10'] = row['GEOID10']
        try:
            df = df[df.enddate.isna()]
        except:
            x.append(index)
        utility_region = utility_region.append(df.loc[:,['GEOID10','utility','eiaid']].drop_duplicates())
    else:
        utility_region = utility_region.append({'region_name': row['GEOID10'],'utility':'No','eiaid':'No'}, ignore_index=True)
```

```python
x
```

```python
utility_region = utility_region.astype({'eiaid':'str'})
utility_region.info()
```

```python
rates = pd.read_excel('../data/rates19/Sales_Ult_Cust_2019.xlsx',
                      sheet_name='States', header=2,usecols='B:E,G:H,K:M', dtype={'Utility Number':'str'}, na_values='.')
rates.info()
```

```python
rates = rates[rates['Megawatthours']>0]
rates['rate'] = rates['Thousand Dollars'] / rates['Megawatthours']
```

```python
rates
```

```python
output = utility_region.merge(rates, left_on='eiaid', right_on='Utility Number')
output
```
